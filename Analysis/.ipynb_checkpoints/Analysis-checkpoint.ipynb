```python
# Pandas and NumPy
import numpy as np
import pandas as pd
# MySQL
import sqlalchemy
```

## Connect to DB

```python
con = sqlalchemy.create_engine("mysql+mysqldb://root:"'1234'"@localhost/secom")
```

## Load data from table_secom

```python
df_table_secom = pd.read_sql('SELECT * FROM table_secom', con, index_col='id',parse_dates='timestamp')
df_table_secom.head()
```

```python
df_weeks = df_table_secom.loc[:,['timestamp','results']]
df_weeks['week'] = df_weeks['timestamp'].dt.week
df_week_behavior = df_weeks[['week', 'results']].groupby(['week']).agg(['count'])
df_week_behavior['count_passed'] = df_weeks[df_weeks['results'] == -1][['week', 'results']].groupby(['week']).agg(['count'])
df_week_behavior['count_failed'] = df_weeks[df_weeks['results'] == 1][['week', 'results']].groupby(['week']).agg(['count'])
df_week_behavior.head()
```
